```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
import datetime
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder

now = datetime.datetime.now()
```

```python
cols = ['a1010', 'a1011', 'a1012', 'a1015', 'a1026', 'a1027', 'a1028','a1029', 'b2010', 'b2017', 'b2031', 'c3012', 'c3013', 'c3014', 'c3016','c3024', 'c3026', 'c3030', 'c3031', 'c3033', 'c3051', 'eau_di', 'd4010','d4011','d4013', 'd4014', 'd4015',
        'd4024', 'd4043', 'd4044', 'd4056','e5020', 'e5039', 'e5090', 'e5091', 'f6013', 'g6010', 'g6012', 'i8019','i8023', 'i8025', 'i8013']

df = pd.read_csv(r'C:\Users\MGUELMOUSSI\Desktop\Dev_projects\CAB-Interface - 50 formules\RR50.csv',sep=';', header=0, index_col=0)
df[cols]=df[cols].fillna(0)
df.shape
```

```python
df_train = df.iloc[:45,:]
df_test = df.iloc[45:,:]
df_test.head(5) 
```

```python
target = 'slump'

X = df_train[cols]
y = df_train[target]

model = xgb.XGBRegressor(colsample_bytree= 0.7, learning_rate= 0.05, max_depth= 5, min_child_weight= 4, n_estimators= 500, nthread= 4, 
objective= 'reg:squarederror', subsample= 0.7 ,booster='gbtree', verbosity=1)
model.fit(X, y)

X_test = df_test[cols]
y_test = df_test[target]


preds = model.predict(X_test)
train_score = model.score(X,y) * 100
test_score = model.score(X_test,y_test) * 100
test_score
```
